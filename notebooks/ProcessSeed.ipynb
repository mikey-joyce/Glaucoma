```python
import os

import pandas as pd
```

```python
data_dir = '../data/seed/'

rename_dict = {
    'bpsys_f_2': 'SBP_2',
    'bpdia_f_2': 'DBP_2',
    'bppul_f_2': 'HR_2'
}

seed = pd.read_csv(data_dir + 'SEED_progression.csv')
seed = seed.rename(columns=rename_dict)

seed.head()
```

```python
def combine_csvs(directory_path):
    dataframes = []

    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        if filename.endswith('.csv') and os.path.isfile(file_path):
            df = pd.read_csv(file_path)
            dataframes.append(df)

    combined_df = pd.concat(dataframes, ignore_index=True)

    return combined_df
```

```python
first = combine_csvs(data_dir + 'SEED_enhanced/')
first = first.drop(['IOP', 'SBP', 'DBP', 'HR'], axis=1)

first.head()
```

```python
temp = combine_csvs(data_dir + 'SEED_enhanced_2/')

second = pd.DataFrame({
    'Patient': temp['Patient'],
    'MAP_2': temp['MAP'],
    'HR_2': temp['HR'],
    'P1_2': temp['P1'],
    'P2_2': temp['P2'],
    'P4_2': temp['P4'],
    'P5_2': temp['P5'],
    'Qmean_2': temp['Qmean'],
    'R1_2': temp['R1'],
    'R4_2': temp['R4'],
    'R5_2': temp['R5']
})

second.head()
```

```python
seed_new = pd.merge(seed, first, on='Patient', how='inner')
seed_new = pd.merge(seed_new, second, on='Patient', how='inner')
seed_new.head()
```

```python
seed_new.to_csv(data_dir + 'SEED_progression_enhanced.csv', index=False)
```
